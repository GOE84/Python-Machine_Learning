```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
import os
print(os.listdir())
```

```python
data = pd.read_csv('train.csv')
```

```python
# 2. แยก "โจทย์ (X)" และ "เฉลย (y)"
y = data.SalePrice
X = data.drop(['SalePrice', 'Id'], axis=1) # ตัด Id ออกเพราะไม่เกี่ยวกับการทำนาย
```

```python
# 3. แบ่งข้อมูลเป็น Train (ไว้สอน) และ Validation (ไว้สอบ)
X_train_full, X_valid_full, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
# เลือกคอลัมน์ตัวเลข
numerical_cols = [cname for cname in X_train_full.columns if X_train_full[cname].dtype in ['int64', 'float64']]
# เลือกคอลัมน์ตัวหนังสือ (เฉพาะที่มีประเภทไม่เยอะมาก เพื่อให้รันเร็ว)
categorical_cols = [cname for cname in X_train_full.columns if X_train_full[cname].dtype == "object" and X_train_full[cname].nunique() < 10]
```

```python
# เลือกเฉพาะคอลัมน์ที่เรากำหนด
my_cols = numerical_cols + categorical_cols
X_train = X_train_full[my_cols].copy()
X_valid = X_valid_full[my_cols].copy()
```

```python
# ตัวจัดการข้อมูลตัวเลข (เติมค่าว่างด้วย Median)
numerical_transformer = SimpleImputer(strategy='median')
```

```python
# ตัวจัดการข้อมูลตัวหนังสือ (เติมค่าว่างด้วยตัวที่เจอบ่อยสุด + แปลงเป็นตัวเลข One-Hot)
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

```python
# --- ส่วนของการสร้างโมเดล ---

# 4. กำหนดโมเดล (ใช้ Random Forest)
model = RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
# สร้าง Pipeline (ท่อส่งงาน: ข้อมูลเข้า -> จัดการข้อมูล -> เข้าโมเดล)
my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', model)])
```

```python
# 5. เริ่มสอนโมเดล (Fit)
print("กำลังสอนโมเดล... รอสักครู่นะครับ")
my_pipeline.fit(X_train, y_train)
```

```python
# 6. ลองทำนายและตรวจคำตอบ
preds = my_pipeline.predict(X_valid)
score = mean_absolute_error(y_valid, preds)

print(f"เรียบร้อย! ค่าความผิดพลาดเฉลี่ย (MAE): {score:,.2f} ดอลลาร์")
```
