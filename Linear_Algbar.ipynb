```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000)
```

**Bank Robber [ d = 2.5t]**

```python
r = 2.5 * t
```

**Sheriff [d = 3 * ( t - 5 )]**

```python
s = 3*(t-5)
```

```python
fig, ax = plt.subplots()
plt.title("A Bank Robber Caught" ,c='black')
plt.xlabel('Time (in minutes)')
plt.ylabel('distance (in KM)')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
```

```python
fig, ax = plt.subplots()
plt.title("A Bank Robber Caught" ,c='black')
plt.xlabel('Time (in minutes)')
plt.ylabel('distance (in KM)')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t,r, color='green')
ax.plot(t,s, color='red')
```
