**สอนคอมพิวเตอร์ให้ทำนายเงินเดือน**

*   numpy: เอาไว้จัดการตัวเลข(คอมพิวเตอร์ชอบเลข)
*   matplotlib: เอาไว้วาดกราฟ (ให้มนุษย์ดูรู้เรื่อง)
*   sklearn (Scikit-learn): พระเอกของเรา คือกล่องเครื่องมือ ML ที่มีสมองสำเร็จรูปให้ใช้







````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
````

**Step 2: เตรียมข้อมูล**
> เราจะสร้างข้อมูลสมมติขึ้นมา (ในงานจริง เราจะโหลดไฟล์ Excel หรือ CSV เข้ามาแทน)

*   X (Features): คือ ข้อมูลที่เรารู้ (ในที่นี้คือ "ปีประสบการณ์")
```
 สังเกต: X ต้องเป็นวงเล็บซ้อน 2 ชั้น [[...]] เพราะ Scikit-learn ต้องการรูปแบบตาราง (2 มิติ)
```

*   y (Target): คือ ข้อมูลที่เราอยากทำนาย (ในที่นี้คือ "เงินเดือน")







````python
x = np.array([[1],[2],[3],[4],[5]])
# สร้างข้อมูลตัวอย่าง (ประสบการณ์ 1-5 ปี)
````

````python
y = np.array([12000,15000,16000,21000,27000])
# เงินเดือนจริงที่สอดคล้องกัน (สมมติว่าเป็นหน่วยบาท)
````

**พล็อตกราฟดูหน้าตาข้อมูลก่อน**


> matplotlib



````python
plt.scatter(x,y, color='red')
````

````python
plt.scatter(x, y, color='red')
plt.title('Experience vs Salary',color='blue')
````

````python
plt.scatter(x, y, color='red')
plt.title('Experience vs Salary',color='blue')
plt.xlabel('Years of Experience' , color='green')
````

````python
plt.scatter(x, y, color='red')
plt.title('Experience vs Salary',color='blue')
plt.xlabel('Years of Experience' , color='green')
plt.ylabel('Salary',color='black')
plt.show()
````

**Step 3: สร้างและสอนโมเดล**



> นี่คือขั้นตอนที่เรียกว่า Machine Learning


*   เราเรียกใช้โมเดล LinearRegression (การถดถอยเชิงเส้น) ซึ่งเหมาะกับการหาความสัมพันธ์ที่เป็นเส้นตรง
*   คำสั่ง .fit(X, y) คือการสั่งให้คอมพิวเตอร์ "เรียนรู้" ว่า X สัมพันธ์กับ y อย่างไร (หาสูตรคณิตศาสตร์ที่เชื่อมโยง 2 ตัวแปรนี้)


```
sklearn (Scikit-learn):
```



````python
model = LinearRegression()
# 1. สร้างโมเดลเปล่าๆ ขึ้นมา
````

````python
model.fit(x,y)
````

**Step 4: ทดสอบทำนาย (Prediction)**


> ตอนนี้โมเดลเราเรียนรู้จบแล้ว เราจะลองถามมันดูว่า "ถ้ามีประสบการณ์ 3.5 ปี และ 7 ปี ควรได้เงินเดือนเท่าไหร่?" (ข้อมูลนี้ไม่มีในตารางสอน)



````python
x_data = np.array([[30],[1]])
# ข้อมูลใหม่ที่ต้องการทำนาย
````

````python
# สั่งทำนาย
prediction = model.predict(x_data)

print(f"ควรได้เงินเดือน X : {prediction[0]:.0f} บาท")
print(f"ควรได้เงินเดือน Y : {prediction[1]:.0f} บาท")
````

**Step 5: ดูเบื้องหลังคณิตศาสตร์**

> สิ่งที่เราทำไปเมื่อกี้ ในทางคณิตศาสตร์คือการหา สมการเส้นตรง

```
y = mx + c
```


*   m (Slope): ความชัน (เพิ่มปีละกี่บาท?)
*   c (Intercept): จุดตัดแกน Y (สตาร์ทเริ่มงาน 0 ปี ได้กี่บาท?)






````python
print(f"ความชัน (m): {model.coef_[0]}")
print(f"จุดตัดแกน Y (c): {model.intercept_}")

# วาดกราฟเส้นที่ AI สร้างขึ้นเทียบกับจุดข้อมูลจริง
plt.scatter(x, y, color='blue', label='Actual Data') # จุดข้อมูลจริง
plt.plot(x, model.predict(x), color='red', label='AI Line') # เส้นที่ AI สร้าง
plt.legend()
plt.show()
````

**สรุปสิ่งที่คุณเพิ่งทำไป**

*   Input: ป้อนข้อมูลประสบการณ์ (X) และเงินเดือน (y)
*   Process: สั่ง model.fit() ให้คอมพิวเตอร์หา pattern (มันไปคำนวณหาค่า m และ c มาให้เราเองโดยที่เราไม่ต้องแก้สมการเอง)
*   Output: สั่ง model.predict() เพื่อทำนายอนาคต
